## Датасет Machine

```python
machine_data = pd.read_csv('../data/machine_preprocessed.csv')
machine_numeric_df = pd.concat([machine_data['myct'], machine_data['mmin'], 
                       machine_data['mmax'],machine_data['cach'],
                       machine_data['chmin'],machine_data['chmax'], machine_data['prp'], machine_data['erp']], axis=1, 
                       keys=['myct', 'mmin','mmax','cach','chmin','chmax','prp','erp'])
machine_numeric_df.hist(bins=20, figsize=(20,20))
plt.show()
```

Сделанная выше визуализация показывает распределение времени машинного цикла, минимального и максимального объема оперативной памяти, кэш-памяти, минимальное и максимальное количество каналов, опубликованной и расчетной производительности относительно общего количества людей в выборке

Из графиков можно сделать следующие выводы. Большая часть ЦП имеют время машинного цикла до 200.
